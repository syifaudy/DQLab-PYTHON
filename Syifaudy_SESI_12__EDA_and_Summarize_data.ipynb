# Import Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ss=pd.read_csv("/content/SuperStore.csv")
```

# Import Dataset

```python
ss.head()
```

#Summary Data

```python
sum_category =pd.pivot_table(
    data=ss,
    index="Category",
    values="Sales",
    aggfunc="sum"
)
sum_category
```

```python
sum_category_2 =pd.pivot_table(
    data=ss,
    index=["Category","Sub-Category"],
    values="Sales",
    aggfunc=["sum","mean"]
)
sum_category_2

#cuma ada 2 kolom
```

```python
sum_category_2.reset_index()

# jadi 4 kolom, (category dan sub category include)
```

```python
ss.head()

#urutan granular paling besar (kedetailan data)

# -customer id
#  -order id
#   -category
#    -sub category
#     - product id
#      - product name
```

```python
popular_product = pd.pivot_table(
    data=ss,
    index="Product_Name",
    values="Order_ID",
    aggfunc="count"
).sort_values(by="Order_ID", ascending=False)

popular_product
```

```python
popular_product.head(10).plot(kind='bar', figsize=(10,5))
```

```python
market_share= pd.pivot_table(
    data=ss,
    index='Category',
    columns='Region',
    values='Sales',
    aggfunc='sum'
)
market_share
```

```python
market_share.plot(kind='bar', figsize=(10,5))
```

```python
market_share.plot(kind='bar', stacked='True', figsize=(10,5))
```

```python
sns.barplot(
    data=ss,
    x='Category',
    y='Sales',
    hue='Region'
)
plt.show()
```

#Line Chart

```python
ss.head()
```

```python
data= pd.pivot_table(
    data=ss,
    index=['Order_Date', 'Category'],
    values='Sales',
    aggfunc='sum'
).reset_index()

data.head()
```

```python
sns.lineplot(data=data, x='Order_Date', y='Sales')
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=data, x='Order_Date', y='Sales', hue='Category')
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(ss, x='Sales', y='Profit', hue='Category')
plt.show()
```

```python
ss['Profit']=ss['Profit'].abs()
sns.scatterplot(ss, x='Sales', y='Profit', hue='Category')
plt.show()
```

```python
sns.distplot(ss['Sales'])
plt.figure(figsize=(2,1))
plt.show()

#datanya skew right
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales',  hue='Category')
plt.show()
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales',  hue='Category', multiple='stack')
plt.show()
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales',  hue='Category', kind='kde')
plt.show()
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales', col='Category', kind='kde')
plt.show()
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales', row='Category', kind='kde')
plt.show()
```

```python
sns.displot(data=ss[ss['Sales']<=1000], x='Sales', col='Region', row='Category', kind='kde')
plt.show()
```
